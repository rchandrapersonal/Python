```python
a1 = 'Hello'
a2 = 'Students'

# Modulo Formatting

print('%s ! Welcome to Edureka %s.' % (a1, a2))
```

```python
# str.format()

print('{}! Welcome to Edureka {}'.format(a1,a2))
```

```python
# f-strings

print(f'{a1}! Welcome to Edureka {a2}')
```

```python
# Template Class
from string import Template

template_var = Template('$n1 ! Welcome to Edureka $n2.')

print(template_var.substitute(n1 = a1, n2=a2))
```

```python
# Template Class  --> String Interpolation

import string

template_var = string.Template('$n1 ! Welcome to Edureka $n2.')

print(template_var.substitute(n1 = a1, n2=a2))
```

#### Mutable Object -- Sets

###### It is a collection of items that are enclosed with curly braces and separated by commas. It has only unique items

```python
var_set = {1, 3, 5, 7, 9, 9}
var_set
```

```python
var_list = [1,2,3,4,5, 7, 7, 8]
var_list
```

```python
var_list = list(set(var_list))
var_list
```

```python
var_set1 = {1,2,3,4,5}
var_set2 = {4,5,6,7,8}
```

```python
# Set Difference
# All elements in var_set1 but not in var_set2

var_set1 - var_set2
```

```python
# All elements in var_set2 but not in var_set1

var_set2 - var_set1
```

```python
# Union of var_set1 and var_set2

var_set1.union(var_set2)
```

```python
# Intersection of var_set1 and var_set2

var_set1.intersection(var_set2)
```
